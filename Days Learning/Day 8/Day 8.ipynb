```python
%sql
-- Create structure
USE CATALOG ecommerce;
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;

-- Register tables
CREATE TABLE bronze.events USING DELTA LOCATION 'workspace.ecommerce.bronze';
CREATE TABLE silver.events USING DELTA LOCATION 'workspace.ecommerce.silver';
CREATE TABLE gold.products USING DELTA LOCATION 'workspace.ecommerce.gold';

-- Permissions
--GRANT SELECT ON TABLE gold.products TO `ref`;
--GRANT ALL PRIVILEGES ON SCHEMA silver TO `engineers@company.com`;
```

```python
%sql
CREATE OR REPLACE VIEW gold.top_products AS
SELECT
  product_id,
  brand,
  category_code,
  COUNT(*)                AS purchases,
  SUM(price)              AS revenue,
  COUNT(DISTINCT user_id) AS unique_buyers
FROM gold.products
WHERE event_type = 'purchase'
GROUP BY product_id, brand, category_code
HAVING COUNT(*) > 10
ORDER BY revenue DESC
LIMIT 100;
```
